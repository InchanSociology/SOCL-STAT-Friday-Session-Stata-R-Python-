```python

# (if you need to install pacakges) use below code after removing #)
# pip install pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

nels = pd.read_stata("C:/Users/InchanHwang/Desktop/Week2/NELS.dta")
```

```python
## Exploring the Dataset##

print(nels.info())
print(nels.head())
```

```python
## Exploring Nominal Variable ##

print(nels['advmath8'].value_counts(dropna=False))
print(nels['region'].value_counts(dropna=False))

## if you want to see missing vairables, use dropna=TRUE

print(pd.crosstab(nels['advmath8'], nels['region'], margins=True))
```

```python
## Two-Way Tabulation/Cross-Tabs ##

print(pd.crosstab(nels['advmath8'], nels['region'], margins=True))
```

```python
## Bar Chart Visualization for Nominal Variables ##

sns.set_theme(style="whitegrid")

nels['advmath8'].value_counts(normalize=True).plot(kind='bar')
plt.title("Percent by advmath8")
plt.show()

nels['region'].value_counts(normalize=True).plot(kind='bar', color='orange')
plt.title("Percent by region")
plt.show()
```

```python
##  Exploring Continuous Variables ##


print(nels['expinc30'].describe())
print("40th percentile:", nels['expinc30'].quantile(0.4))

print(nels.groupby('region')['expinc30'].describe())
```

```python
# histogram


s = nels['expinc30'].dropna()
mu, sigma = s.mean(), s.std()

sns.histplot(s, bins=30, stat='density')
x = np.linspace(s.min(), s.max(), 400)
plt.plot(x, (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-0.5*((x-mu)/sigma)**2), color='red')
plt.title("Histogram with Normal Curve")
plt.show()

# histogram across different region
sns.displot(nels, x='expinc30', col='region', bins=30, stat='density')
plt.show()

# if (<250000)
subset = nels[nels['expinc30'] < 250000]['expinc30']
sns.histplot(subset, bins=30, stat='density')
plt.title("Histogram (<250000)")
plt.show()

```

```python
# Boxplot

sns.boxplot(data=nels, x='region', y='expinc30', color='#BCBDDC')
plt.title("Boxplot by region")
plt.show()

sns.boxplot(data=nels[nels['expinc30'] < 250000], x='region', y='expinc30', color='#BFD3E6')
plt.title("Boxplot (<250000)")
plt.show()
```
